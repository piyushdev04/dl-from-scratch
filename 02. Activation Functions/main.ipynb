```python
import numpy as np
import matplotlib.pyplot as plt

# activation functions

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def gelu(x):
    return 0.5 * x * (1 + np.tanh(
        np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)
    ))

def softmax(z):
    e = np.exp(z - np.max(z))
    return e / np.sum(e)

# input range

x = np.linspace(-6, 6, 500)

# global style

plt.rcParams.update({
    "font.size": 11,
    "axes.spines.top": False,
    "axes.spines.right": False
})

# plot 

fig, axs = plt.subplots(2, 3, figsize=(15, 9))
fig.suptitle("Activation Functions", fontsize=18)

# Sigmoid
axs[0,0].plot(x, sigmoid(x), linewidth=3)
axs[0,0].axhline(0, alpha=0.2)
axs[0,0].axhline(1, alpha=0.2)
axs[0,0].set_title("Sigmoid")
axs[0,0].set_ylim(-0.1,1.1)
axs[0,0].grid(alpha=0.3)

# Tanh
axs[0,1].plot(x, tanh(x), linewidth=3)
axs[0,1].axhline(0, alpha=0.3)
axs[0,1].set_title("Tanh")
axs[0,1].set_ylim(-1.1, 1.1)
axs[0,1].grid(alpha=0.3)

# ReLU
axs[0,2].plot(x, relu(x), linewidth=3)
axs[0,2].axhline(0, alpha=0.3)
axs[0,2].set_title("ReLU")
axs[0,2].set_ylim(-1, 6)
axs[0,2].grid(alpha=0.3)

# GELU
axs[1,0].plot(x, gelu(x), linewidth=3)
axs[1,0].axhline(0, alpha=0.3)
axs[1,0].set_title("GELU")
axs[1,0].grid(alpha=0.3)

# Softmax (bar style)
z = np.array([-3, -2, -1, 0, 1, 2, 3])
axs[1,1].bar(z, softmax(z))
axs[1,1].set_title("Softmax (Probabilities)")
axs[1,1].set_xlabel("Input Score")
axs[1,1].set_ylabel("Probability")
axs[1,1].grid(axis="y", alpha=0.3)

# hide empty subplot
axs[1,2].axis("off")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
