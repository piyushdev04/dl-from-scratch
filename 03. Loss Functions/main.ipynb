```python
import numpy as np
import matplotlib.pyplot as plt

# loss functions

def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

def cross_entropy(y_true, y_pred):
    eps = 1e-10
    y_pred = np.clip(y_pred, eps, 1 - eps)
    return -np.mean(
        y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred)
    )

# visualization setup

plt.rcParams.update({
    "font.size": 11,
    "axes.spines.top": False,
    "axes.spines.right": False
})

fig, axs = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Loss Functions Visualization", fontsize=16)

# 1. MSE visualization
y_true = 1
y_preds = np.linspace(-1, 3, 300)
mse_vals = [(y_true - yp)**2 for yp in y_preds]

axs[0].plot(y_preds, mse_vals, linewidth=3)
axs[0].axvline(y_true, linestyle="--", alpha=0.5)
axs[0].set_title("Mean Squared Error (Regression)")
axs[0].set_xlabel("Prediction")
axs[0].set_ylabel("Loss")
axs[0].grid(alpha=0.3)

# 2. Cross-Entropy Visualization
probs = np.linspace(0.001, 0.999, 300)
ce_vals = [-np.log(p) for p in probs]

axs[1].plot(probs, ce_vals, linewidth=3)
axs[1].axvline(1, linestyle="--", alpha=0.5)
axs[1].set_title("Cross-Entropy Loss (Classification)")
axs[1].set_xlabel("Predicted Probability (True Class)")
axs[1].set_ylabel("Loss")
axs[1].grid(alpha=0.3)

plt.tight_layout(rect=[0, 0, 1, 0.92])
plt.show()
```
