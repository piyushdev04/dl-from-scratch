```python
import numpy as np
import matplotlib.pyplot as plt

# generate simple data
# y = 2x + noise

np.random.seed(42)
X = np.linspace(-3, 3, 100).reshape(-1, 1)
y = 2 * X + np.random.randn(100, 1) * 0.5

# initialize parameters

w = np.random.randn()
b = 0.0
lr = 0.05
epochs = 100

losses = []
grad_w = []

# training loop (backprop)
for epoch in range(epochs):

    # Forward pass
    y_pred = w * X + b
    loss = np.mean((y - y_pred) ** 2)
    losses.append(loss)

    # Backward pass (gradients)
    dw = -2 * np.mean(X * (y - y_pred))
    db = -2 * np.mean(y - y_pred)

    grad_w.append(dw)

    # UPDATE
    w -= lr * dw
    b -= lr * db

    if epoch % 10 == 0:
        print(f"Epoch {epoch} | Loss: {loss:.4f}")

print("Final w:", w)
print("Final b:", b)
```

```python
# visualization - Loss Curve

plt.figure(figsize=(7,4))
plt.plot(losses, linewidth=3)
plt.xlabel("Epoch")
plt.ylabel("Loss (MSE)")
plt.title("Loss Decreasing via Backpropagation")
plt.grid(alpha=0.3)
plt.show()
```

```python
# visualization - gradient flow

plt.figure(figsize=(7,4))
plt.plot(grad_w, linewidth=3)
plt.axhline(0, linestyle="--", alpha=0.4)
plt.xlabel("Epoch")
plt.ylabel("Gradient (dw)")
plt.title("Gradient Flow During Backpropagation")
plt.grid(alpha=0.3)
plt.show()
```
