```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Pooling From Scratch
def pooling2d(feature_map, pool_size=2, stride=2, mode="max"):
    h, w = feature_map.shape
    out_h = (h - pool_size) // stride + 1
    out_w = (w - pool_size) // stride + 1

    output = np.zeros((out_h, out_w))

    for i in range(out_h):
        for j in range(out_w):
            region = feature_map[
                i*stride : i*stride + pool_size,
                j*stride : j*stride + pool_size
            ]

            if mode == "max":
                output[i, j] = np.max(region)
            elif mode == "avg":
                output[i, j] = np.mean(region)

    return output

# 2. Sample Feature Map
np.random.seed(42)
feature_map = np.random.rand(8, 8)

# 3. Apply Pooling
max_pooled = pooling2d(feature_map, pool_size=2, stride=2, mode="max")
avg_pooled = pooling2d(feature_map, pool_size=2, stride=2, mode="avg")

# 4. visualization
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(feature_map, cmap="viridis")
plt.title("Input Feature Map")
plt.colorbar()

plt.subplot(1,3,2)
plt.imshow(max_pooled, cmap="coolwarm")
plt.title("Max Pooling (2x2)")
plt.colorbar()

plt.subplot(1,3,3)
plt.imshow(avg_pooled, cmap="coolwarm")
plt.title("Average Pooling (2x2)")
plt.colorbar()

plt.tight_layout()
plt.show()
```
