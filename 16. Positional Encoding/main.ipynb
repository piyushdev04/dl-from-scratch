```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Positional Encoding Function
def positional_encoding(seq_len, d_model):
    PE = np.zeros((seq_len, d_model))

    for pos in range(seq_len):
        for i in range(0, d_model, 2):
            div_term = np.exp(-np.log(10000.0) * (i / d_model))
            PE[pos, i] = np.sin(pos * div_term)
            if i + 1 < d_model:
                PE[pos, i + 1] = np.cos(pos * div_term)

    return PE

# 2. Generate Positional Encoding
seq_len = 20
d_model = 16

PE = positional_encoding(seq_len, d_model)

print("Positional Encoding Shape:", PE.shape)
print("\nFirst 3 Positions:\n", PE[:3])

# 3. Visualization (Heatmap)
plt.figure(figsize=(10, 5))

plt.imshow(PE, cmap="viridis", aspect="auto")
plt.colorbar(label="Encoding Value")

plt.xlabel("Embedding Dimension")
plt.ylabel("Position (Time Step)")
plt.title("Sinusoidal Positional Encoding")

plt.xticks(range(d_model))
plt.yticks(range(seq_len))

plt.grid(False)
plt.tight_layout()
plt.show()
```
